```julia
using PyPlot
const plt = PyPlot
# ↑Python の import matplotlib.pyplot as plt 相当
```

```julia
x = 0:0.1:6
y = sin.(x);
```

※↑行末が `;` で終わる行でセルが終わると、結果を出力しない（豆）

```julia
plt.plot(x, y)
```

```julia
y1 = sin.(x);
y2 = cos.(x);
```

```julia
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
```
